#### Popularity-Based Filtering (Global Ranking)

Popularity-Based Filtering (Global Ranking): This method ranks items based on global metrics (like the Weighted Rating above). It doesn't use any user information other than their ratings for the items themselves. It's often the simplest baseline model.

Demographic Filtering: This is a slightly more specific approach that involves segmenting users based on shared characteristics (e.g., age, location, gender, profession) and then recommending items that are popular within that specific demographic. The $\text{WR}$ can be applied here too, but the $R$ and $C$ values would be calculated only for users within the target demographic.

```python
# IMDB weighted rating formula
# W = (Rv + Cm) / (v + m)
# where,
# R = average rating of the movie
# v = number of votes for the movie
# m = minimum votes required to be listed in the chart
# C = the mean vote across the whole reported dataset   


# How it Works:Low Votes ($v \ll m$): 
# If an item has few votes, the first term $\left(\frac{v}{v+m} \cdot R\right)$ is small, and the second term $\left(\frac{m}{v+m} \cdot C\right)$ dominates. 
# This effectively pulls the item's final rank toward the global average ($C$), penalizing it.

# High Votes ($v \approx m$): As the number of votes ($v$) approaches the threshold ($m$), the weight shifts, 
# and the item's actual average rating ($R$) begins to influence the $\text{WR}$ much more.

# Very High Votes ($v \gg m$): If an item has many votes, the first term approaches $1 \cdot R$, 
# and the weighted rating essentially equals the item's average rating ($R$), confirming its true popularity and quality.
```

```python
C= df2['vote_average'].mean()
C
```

```python
# We use 90th percentile as our cutoff. In other words, for a movie to feature in the charts, 
# it must have more votes than at least 90% of the movies in the list
m= df2['vote_count'].quantile(0.9)
m
```

```python
# filter out the movies that qualify for the chart
q_movies = df2.copy().loc[df2['vote_count'] >= m]
q_movies.shape
```

```python
# penalize items with very few ratings using the weighted average for ranking

def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# Define a new feature 'score' and calculate its value with `weighted_rating()`
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
#Sort movies based on score calculated above
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```
